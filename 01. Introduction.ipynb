```python
!pip install matplotlib
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Training data
X = [[1], [2], [3], [4]]
y = [2, 4, 6, 8]

# Create and train model
model = LinearRegression()
model.fit(X, y)

# Predict values
y_pred = model.predict(X)

# Plot data points
plt.scatter(X, y, label="Actual Data")

# Plot regression line
plt.plot(X, y_pred, label="Regression Line")

plt.xlabel("X")
plt.ylabel("y")
plt.title("Simple Linear Regression")
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

X = [[1, 1], [2, 1], [3, 2], [4, 3],
     [1, 3], [2, 4], [3, 5], [4, 6]]
y = [0, 0, 0, 0, 1, 1, 1, 1]

model = LogisticRegression()
model.fit(X, y)

plt.scatter([x[0] for x in X], [x[1] for x in X], c=y)
plt.title("Simple Classification")
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import numpy as np

# Simple 2D data
X = [[1, 1], [2, 1], [3, 2], [4, 3],
     [1, 3], [2, 4], [3, 5], [4, 6]]

y = [0, 0, 0, 0, 1, 1, 1, 1]   # Class labels

# Create and train model
model = DecisionTreeClassifier()
model.fit(X, y)

# Separate points by class for plotting
X0 = [x[0] for x, label in zip(X, y) if label == 0]
Y0 = [x[1] for x, label in zip(X, y) if label == 0]

X1 = [x[0] for x, label in zip(X, y) if label == 1]
Y1 = [x[1] for x, label in zip(X, y) if label == 1]

# Plot data points
plt.scatter(X0, Y0, label="Class 0")
plt.scatter(X1, Y1, label="Class 1")

# Create a grid to show decision regions
xx, yy = np.meshgrid(np.linspace(0, 5, 100),
                     np.linspace(0, 7, 100))
grid = np.c_[xx.ravel(), yy.ravel()]
Z = model.predict(grid)
Z = Z.reshape(xx.shape)

# Plot decision regions
plt.contourf(xx, yy, Z, alpha=0.3)

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Simple Classification using Decision Tree")
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Simple 2D data
X = [[1, 2], [1, 4], [1, 0],
     [10, 2], [10, 4], [10, 0]]

# Create K-Means model (2 clusters)
model = KMeans(n_clusters=2, random_state=0)
model.fit(X)

# Get cluster labels
labels = model.labels_

# Plot data points colored by cluster
x_values = [x[0] for x in X]
y_values = [x[1] for x in X]

plt.scatter(x_values, y_values, c=labels)
plt.title("Simple Clustering using K-Means")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
